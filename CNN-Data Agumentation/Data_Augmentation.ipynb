## Data Agumentation

```
# Load all the Library
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
import cv2
import numpy as np
import tensorflow as tf

# load the image
img = load_img('bird.jpg')
pyplot.imshow(img)
```

```
image = np.array(img)
print(image.shape)

img = cv2.resize(image, (500,500))
pyplot.imshow(img)
```

```
# expand dimension to one sample
samples = expand_dims(img, 0)
# create image data augmentation generator
datagen = ImageDataGenerator(width_shift_range=[-200,200])
# prepare iterator
it = datagen.flow(samples, batch_size=2)
# generate samples and plot
fig=pyplot.figure(figsize=(9, 9))

for i in range(9):
	# define subplot
	fig.add_subplot(330 + 1 + i)
	# generate batch of images
	batch = it.next()
	# convert to unsigned integers for viewing
	image = batch[0].astype('uint8')
	# plot raw pixel data
	pyplot.imshow(image)
# show the figure
pyplot.show()
```

# Random Brightness Augmentation
##### Values less than 1.0 darken the image, e.g. [0.5, 1.0], whereas values larger than 1.0 brighten the image, e.g. [1.0, 1.5], where 1.0 has no effect on brightness.

```
# expand dimension to one sample
samples = expand_dims(img, 0)
# create image data augmentation generator
datagen = ImageDataGenerator(brightness_range=[2,1])
# prepare iterator
it = datagen.flow(samples, batch_size=2)
# generate samples and plot
fig=pyplot.figure(figsize=(9, 9))

for i in range(9):
	# define subplot
	fig.add_subplot(330 + 1 + i)
	# generate batch of images
	batch = it.next()
	# convert to unsigned integers for viewing
	image = batch[0].astype('uint8')
	# plot raw pixel data
	pyplot.imshow(image)
# show the figure
pyplot.show()
```

# Random Zoom Augmentation:
Image zooming can be configured by the zoom_range argument to the ImageDataGenerator constructor. You can specify the percentage of the zoom as a single float or a range as an array or tuple.

If a float is specified, then the range for the zoom will be [1-value, 1+value]. For example, if you specify 0.3, then the range will be [0.7, 1.3], or between 70% (zoom in) and 130% (zoom out).

The zoom amount is uniformly randomly sampled from the zoom region for each dimension (width, height) separately.

The zoom may not feel intuitive. Note that zoom values less than 1.0 will zoom the image in, e.g. [0.5,0.5] makes the object in the image 50% larger or closer, and values larger than 1.0 will zoom the image out by 50%, e.g. [1.5, 1.5] makes the object in the image smaller or further away. A zoom of [1.0,1.0] has no effect.

```
# expand dimension to one sample
samples = expand_dims(img, 0)
# create image data augmentation generator
datagen = ImageDataGenerator(zoom_range=[0.5,1.0])
# prepare iterator
it = datagen.flow(samples, batch_size=2)
# generate samples and plot
fig=pyplot.figure(figsize=(9, 9))

for i in range(9):
	# define subplot
	fig.add_subplot(330 + 1 + i)
	# generate batch of images
	batch = it.next()
	# convert to unsigned integers for viewing
	image = batch[0].astype('uint8')
	# plot raw pixel data
	pyplot.imshow(image)
# show the figure
pyplot.show()
```

# ZCA Whitening
A whitening transform of an image is a linear algebra operation that reduces the redundancy in the matrix of pixel images.

Less redundancy in the image is intended to better highlight the structures and features in the image to the learning algorithm.

Typically, image whitening is performed using the Principal Component Analysis (PCA) technique. More recently, an alternative called ZCA (learn more in Appendix A of this tech report) shows better results and results in transformed images that keeps all of the original dimensions and unlike PCA, resulting transformed images still look like their originals.

You can perform a ZCA whitening transform by setting the zca_whitening argument to True.

```
# expand dimension to one sample
samples = expand_dims(img, 0)
# create image data augmentation generator
datagen = ImageDataGenerator(zca_whitening=True)
# prepare iterator
it = datagen.flow(samples, batch_size=2)
# generate samples and plot
fig=pyplot.figure(figsize=(9, 9))

for i in range(9):
	# define subplot
	fig.add_subplot(330 + 1 + i)
	# generate batch of images
	batch = it.next()
	# convert to unsigned integers for viewing
	image = batch[0].astype('uint8')
	# plot raw pixel data
	pyplot.imshow(image)
# show the figure
pyplot.show()
```
